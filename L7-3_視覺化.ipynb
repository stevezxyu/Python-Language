```python
# Matplotlib
# Python 基礎的繪圖套件
#  %matplotlib inline # 讓圖能夠在 Notebook 中顯示，不一定需要
# import matplotlib.pyplot as plt

'''
向 Matplotlib 說嗨
圖、子圖
不同的圖形
'''

# 使用預設值
import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
plt.plot(x, cos_x)
plt.plot(x, sin_x)

# 顯示
plt.show()      # 全部作完圖在show出來
```

```python
# 試著更換線的寬度與顏色
import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
plt.plot(x, cos_x, color = "blue", linewidth = 3, linestyle = ":")    # 更改線的顏色、寬度、類型
plt.plot(x, sin_x, color = "red", linewidth = 3, linestyle = "--")

# 顯示
plt.show()
```

```python
# 試著更換 X 軸與 Y 軸的範圍
import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
plt.plot(x, cos_x, color = "blue", linewidth = 3, linestyle = "-")
plt.plot(x, sin_x, color = "red", linewidth = 3, linestyle = "-")
plt.xlim(x.min()*2, x.max()*2)                # x 軸範圍的最大值, 最小值更改為 2倍
plt.ylim(cos_x.min()*1.2, cos_x.max()*1.2)    # y 軸範圍的最大值, 最小值更改為 1.2倍

# 顯示
plt.show()
```

```python
# 試著更換座標軸刻度
# 試著更換座標軸刻度的顯示標籤

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
plt.plot(x, cos_x, color = "blue", linewidth = 3, linestyle = "-")
plt.plot(x, sin_x, color = "red", linewidth = 3, linestyle = "-")
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(cos_x.min()*1.2, cos_x.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'])
# 更改 x軸的刻度, 且加上標籤 label

plt.yticks([-1, 0, 1], ['$-1$', '$0$', '$+1$'])    # 更改 y軸的刻度, 且加上標籤 label

# 顯示
plt.show()
```

```python
# 試著將圖的外框（spine）移動
# 隱藏一個水平、一個垂直
# 將剩餘的兩個移動至原點

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
# 加上標籤
plt.plot(x, cos_x, color = "blue", linewidth = 3, linestyle = ":", label = "cos(x)" )
plt.plot(x, sin_x, color = "red", linewidth = 3, linestyle = "--", label = "sin(x)")
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(cos_x.min()*1.2, cos_x.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
plt.yticks([-1, 0, 1], ["$-1$", "$0$", "$+1$"])

ax = plt.gca()          # get current axes
ax.spines['right'].set_color('none')     # 水平刻度右移到邊線

ax.spines['top'].set_color('none')       # 垂直刻度上移到邊線

#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))   # 把資料的標籤往下移

#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))     # 把資料的標籤往左移

# 增加圖例
plt.legend(loc = 'upper left', frameon = False)    # loc 位置, frameon 外框

# 顯示
plt.show()
```

```python
import pandas as pd

url = "https://storage.googleapis.com/py_ml_datasets/aapl.csv"
data = pd.read_csv(url, index_col = 0, parse_dates = True)
url_m = "https://storage.googleapis.com/py_ml_datasets/msft.csv"
msft = pd.read_csv(url_m, index_col = 0, parse_dates = True)

plt.plot_date(x = data.index, y = data['Close'].values, fmt = "r-", label = "APPL")
plt.plot_date(x = msft.index, y = msft['Close'].values, fmt = "b-", label = "MSFT")
plt.legend(loc = 'center right', frameon = False)       # 放在中心的右方
plt.show()
```

```python
'''
plt.figure()
figsize 是圖片的長、寬（吋）
dpi 是圖片的解析度（Dots per inch）
'''

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

plt.figure()     # 試著加入 figsize 與 dpi 參數調整看看
plt.plot(x, cos_x)
plt.plot(x, sin_x)
plt.show()
```

```python
'''
plt.subplot()
列數
欄數
第幾個圖
'''

import matplotlib.pyplot as plt

plt.subplot(2,1,1)
plt.xticks([])   # 隱藏 X軸的刻度

plt.subplot(2,3,4)
plt.yticks([])   # 隱藏 y軸的刻度
plt.subplot(2,3,5)
plt.yticks([])
plt.subplot(2,3,6)
plt.yticks([])

plt.show()
```

```python
# 運用 gridspec 進行更進階的子圖切割

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

G = gridspec.GridSpec(3, 3)       # 指定切割圖形

plt.subplot(G[0, :])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot 1', ha = 'center',va = 'center', size = 18)

plt.subplot(G[1, :2])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot 2', ha = 'center',va = 'center', size = 18)

plt.subplot(G[1:, 2])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot 3', ha = 'center',va = 'center', size = 18)

plt.subplot(G[2, 0])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot 4', ha = 'center',va = 'center', size = 18)

plt.subplot(G[2, 1])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot 5', ha = 'center',va = 'center', size = 18)

plt.show()
```
